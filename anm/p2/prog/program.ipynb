```julia
# P(x) = x^5 - 7x^3+ 0.5x^2+1
# P'(x)= 5x^4 - 21x^2 + x
# P(1) = 1 - 7 + 0.5 + 1 = -4.5
myplot(clenshawnorton((x,y) -> 5x^4 - 21x^2 + x, -4.5, 1, 4), :topright)
```

```julia
# f(x) = 0.5e^(2x)
# f'(x)= e^(2x)
# f(1) = 0.5e^2
e = MathConstants.e
myplot(clenshawnorton((x,y) -> e^(2x), 0.5*e^2, 1, 5), :topleft)
```

```julia
# y = 1/(x^3-7)
myplot(clenshawnorton((x,y) -> -3*x^2*y^2, 1/8, -1, 6))
```

```julia
# y' = sqrt(y^2+1)x, y(-1) = 0, y(x) = sinh(1/2(x^2-1))
myplot(clenshawnorton((x,y) -> sqrt(y^2+1)*x, 0.0, -1, 6), :topright)
```
